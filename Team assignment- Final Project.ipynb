## MISSING VALUE & OUTLIERS 
1. MISSING VALUES

Regarding missing values, if we assume the existence of the “true” values for the missing ones, the known information could be used to estimate these missing values. However, the simplest approach that we use for utilizing data-dependent information is to replace unknown values by mean/mode/median. (Wohlrab & Fürnkranz, 2011).

The common option is using mean as the "representative" of the missing feature better for continuous/numeric attribute no matter how many times we add mean, it still gets conserved. However, especially the number of missing values is large, there are a couple of problems with mean, it is heavily influenced by outliers and reduces the variance of the data, or even though mean can be represented for the feature, imputing missing values by mean can worsen the central tendency of normal data. (Kubara, 2019)

To solve the tendency problem, replacing missing values by the mode is the best approach. However, ambiguities can be created when it comes to continuous data because more than one mode or none of all could be found. Additionally, the mean and mode replacement will ignore the feature correlation which is a bad approach for the data. (Badr, 2019) 

Therefore, our team decided to use the median since it is not influenced by the datasets outliers (Badr, 2019).. We grouped the data by the region and calculated the median: each missing value for each country was imputed by the median of its region. The type of world bank is a survey which means the known features from outside sources are not reliable due to the differences between years when data were collected. There were some leftover missing values (because some regions have a lot of missing values in this column), but we accepted the result and ignore the leftovers. 

After imputing all the missing values, we tested both original and imputed data graphs: since there was no big differences between 2 datasets, we moved forward and filtered South America from imputed data to further step in the outliers investigation. 

2. OUTLIERS

Identifying the outliers is a crucial part of statistical analysis, these extreme values are different from the majority of data which has a substantial influence on the analysis and could cause distortion or inaccurate conclusions. There are 3 categories: point outliers (simplest one), contextual outliers, and collective outliers (R. Bansal et al., 2016). The reasons causing outliers in datasets are data entry errors by human mistakes, measurement errors, experimental errors, sampling errors or data manipulation/data set unintended mutations, etc. (Santoyo, 2017).

On our report, we discussed about 2 common techniques to set threshold such as: 
- Z-score: a simple but powerful method to detect the outliers in the dataset. Some good ‘thumb-rule’ thresholds can be 2.5, 3, or more standard deviation far away from the Z-score (0), the further away shows the more unusual the feature is. However, if the data is not the normal distribution, the z-score method might not be accurate. (Santoyo, 2017)
- The Interquartile Range Rule: it sets the threshold and box-plot to visualize and detect outliers. The box-plot graph uses the interquartile method with fences to find outliers. When visualizing the box-plot, the outliers are defined as a data point that is located outside the whiskers of the box-plot (“5 Ways to Find Outliers in Your Data,” 2019). To set our threshold, we used the IQR technique to calculate cut off points that are outside the first quartile (25th) and the third quartile (75th).
